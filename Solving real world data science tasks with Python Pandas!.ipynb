```python
all_data=pd.read_csv("all_data.csv")
```

```python
nan_df=all_data[all_data.isna()]
nan_df
```

```python
all_data.dropna(how="all",inplace=True)
all_data
```

```python
all_data["months"]=all_data["Order Date"].str[0:2]
pd.set_option("display.max_rows",100)
all_data.head(100)
```

```python
all_data["months"].value_counts()
```

Find  or value and delete it

```python
temp_df=all_data[all_data["months"]=="Or"]
temp_df
```

```python
all_data=all_data[all_data["months"]!="Or"]
all_data["months"]=pd.to_numeric(all_data["months"])
all_data["Quantity Ordered"]=pd.to_numeric(all_data["months"])
all_data["Price Each"]=pd.to_numeric(all_data["Price Each"])
all_data.info()
```

```python
all_data["months"].value_counts()
```

Find sales

```python
all_data["sales"]=all_data["Price Each"]*all_data["Quantity Ordered"]
all_data.head()
```

BEST MONTH FOR SALES

```python
x=all_data.groupby("months").sum()
x
```

```python

fig, axes=plt.subplots(2,figsize=(20,11))

axes[0].bar(x.index, x["sales"],color="darkblue")
axes[0].set_xticks(x.index)
axes[0].set_xlabel("Months",fontsize=15)
axes[0].set_ylabel("Salesin USD", fontsize=15) 

axes[1].hist(all_data["months"], edgecolor="black")
axes[1].set_xticks(x.index) 
axes[1].set_xlabel("Months",fontsize=15)
axes[1].set_ylabel("Number of products sold", fontsize=15) 



axes[0].grid()
axes[1].grid()
```

```python
def get_city(address):
    return address.split(",")[1]

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data["city"]=all_data["Purchase Address"].apply( lambda x: f"{get_city(x)} ({get_state(x)})")

```

```python
all_data.head()
```

```python
results=all_data.groupby("city").sum()
results[['sales']]
```

```python
fig,axes=plt.subplots(2,figsize=(15,9))

city=(x for x, core in results)

axes[0].bar(results.index, results["sales"])
axes[0].set_xticklabels(results.index,rotation=45)
axes[0].set_xlabel("City", fontsize=20)
axes[0].set_ylabel("Sales in USD $",fontsize=15)

axes[1].hist(all_data["city"], edgecolor="black" )
axes[1].set_xticklabels(results.index)
axes[1].set_xlabel("City", fontsize=20)
axes[1].set_ylabel("Number of product sold",fontsize=15)

fig.tight_layout(pad=5.0)
```

Time to display advertisment to maximize likelihood of customer's buying product 

```python
all_data["Order Date"]=pd.to_datetime(all_data["Order Date"])
all_data["Hour"]=all_data["Order Date"].dt.hour
all_data["Minute"]=all_data["Order Date"].dt.minute
```

```python
all_data.head()
```

```python
plt.figure(figsize=(17,7))
sales=all_data.groupby(["Hour"]).count()

hours=(key for key, df in sales)

plt.plot(sales.index,sales)
plt.xticks(sales.index)
plt.xlabel("Hours", fontsize=15)
plt.ylabel("Numbers of orders placed ", fontsize=15)
plt.plot()
plt.grid()
```

Products which are sold often together

```python
df=all_data[all_data["Order ID"].duplicated(keep=False)]
df["Grouped"]=df.groupby("Order ID")[["Product"]].transform(lambda x: ", ".join(x))
df=df[["Order ID","Grouped" ]].drop_duplicates()
df
```

```python
from itertools import combinations
from collections import Counter

count=Counter()

for row in df["Grouped"]:
    row_list=row.split(",")
    count.update(Counter(combinations(row_list,2)))

print(count)
```

```python
count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```
